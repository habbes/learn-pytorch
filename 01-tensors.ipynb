# Tensors

Based on https://docs.pytorch.org/tutorials/beginner/blitz/tensor_tutorial.html

```python
import torch
import numpy as np
```

## Tensor initialization

```python
# Directly from data / Python data structures
data = [[1,2], [3,4]]
x_data = torch.tensor(data)
```

```python
# From numpy array
np_array = np.array(data)
```

```python
# From anoter tensor
x_ones = torch.ones_like(x_data) # returns properties of x_data, i.e. shape and datatype
print(f"Ones Tensors: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

```python
# with random or constant values

shape = (2, 3)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

## Tensor Attributes

```python
tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

## Tensor Operations

Tensor operations reference: https://pytorch.org/docs/stable/torch.html

```python
# We move our tensor to GPU if available
if torch.cuda.is_available():
    tensor = tensor.to('cuda')
    print(f"Device tensor is stored on: {tensor.device}")
```

```python
# numpy-like indexing and slicing
tensor = torch.ones(4, 4)
tensor[:, 1] = 0
print(tensor)
```

```python
# concatenate tensors with torch.cat

t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)

t2 = torch.cat([tensor, tensor, tensor], dim=0)
print(t2)
```

```python
# multiplying tensors (element-wise product)
print(f"tensor.mul(tensor): \n {tensor.mul(tensor)} \n")
# equivalent to
print(f"tensor * tensor: \n {tensor * tensor}")
```

```python
# matrix-matrix multiplication

print(f"tensor.matmul(tensor.T): \n {tensor.matmul(tensor.T)} \n")
# equivalent to
print(f"tensor @ tensor.T: \n {tensor @ tensor.T}")
```

```python
# Operations that have a _ suffix are in-place, e.g. x.copy_(y), x.t_(), will change x itself.

print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

## Bridge with numpy

```python
# tensor to numpy array

t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

```python
# A change in the tensor reflects in the NumPy array

t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

```python
# numpy to tensor

n = np.ones(5)
t = torch.from_numpy(n)
```

```python
# Changes in the Numpy array reflects in the tensor

np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
